```python
from tensorflow import keras
from tensorflow.keras import layers

# Criando o neurônio
modelo = keras.Sequential([
    layers.Dense(units = 1, input_shape = [11]) # units é a saída (queremos saber a qualidade do vinho) e o input_shape são os dados de entrada (todas as outras informações sobre o vinho que não seja o dado que queremos saber)
])
```

```python
w, b = modelo.weights

print("Weights\n{}\n\nBias\n{}".format(w, b))
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt

# Modelo de 1 entrada e 1 saída
model = keras.Sequential([
    layers.Dense(1, input_shape=[1]),
])

# em teoria vamos ter um gráfico em linha reta com valores aleatórios pois o modelo ainda não foi treinado
x = tf.linspace(-1.0, 1.0, 100)
y = model.predict(x)

plt.figure(dpi=100)
plt.plot(x, y, 'k')
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.xlabel("Input: x")
plt.ylabel("Target y")
w, b = model.weights
plt.title("Weight: {:0.2f}\nBias: {:0.2f}".format(w[0][0], b[0]))
plt.show()
```
