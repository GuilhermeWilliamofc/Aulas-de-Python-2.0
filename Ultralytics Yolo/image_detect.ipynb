# Detecção de Imagem
Esse Script usa o modelo Yolo11x para detectar elementos em uma imagem

```python
from ultralytics import YOLO

# Carregar o modelo YOLO
# Certifique-se de que o arquivo 'yolo11n.pt' está no caminho correto
model = YOLO('models\yolo11x.pt')

# Realizar a predição na imagem
results = model('images\city.jpg', save=True, show=True)

# Mostrar os resultados da primeira imagem (se houver mais de uma)
results[0].show()
```
